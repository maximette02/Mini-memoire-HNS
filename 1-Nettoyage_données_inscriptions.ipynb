Nettoyage de la colonne Place

```python
import pandas as pd

df_inscriptions = pd.read_csv('/content/cyprus-inscriptions.csv', sep=',', encoding='utf-8-sig', engine='python')
df_inscriptions
```

```python
# Nettoyer le DataFrame pour les noms de lieux
split_1 = df_inscriptions['Place'].str.split(':', n=1, expand=True)
df_inscriptions['Place_Region'] = split_1[0].str.strip()
df_inscriptions['Place_Site_Raw'] = split_1[1].str.strip()
split_2 = df_inscriptions['Place_Site_Raw'].str.split(r'[\(\"]', n=1, expand=True)
df_inscriptions['Place_Site'] = split_2[0].str.strip()
df_inscriptions['Place_precision'] = split_2[1].str.replace(r'[\)\"]', '', regex=True).str.strip()
df_inscriptions.drop(columns=['Place_Site_Raw'], inplace=True)
df_inscriptions.to_csv('cyprus-inscriptions-clean.csv', index=False)
print(df_inscriptions.value_counts('Place_Region'))
print(df_inscriptions.value_counts('Place_Site'))
print(df_inscriptions.value_counts('Place_precision'))
```

On voit qu'il n'y a pas énormément de noms de région, ce qui est bien pour nous. Il conviendra de nettoyer quelques valeurs. La pratique rigoureuse sera de drop les valeurs avec un point d'interrogation du DataFrame.
Quant aux sites, j'affiche les valeurs uniques pour voir si on ne peut pas les normaliser.

```python
df_inscriptions = df_inscriptions[~df_inscriptions['Place'].str.contains(r'\?', na=False)]
df_inscriptions.value_counts('Place_Region')
```

```python
df_inscriptions['Place_Site'].unique()
```

```python
df_inscriptions['Place_Site'] = df_inscriptions['Place_Site'].str.replace(r'[\r\n]+', ' ', regex=True).str.strip()
df_inscriptions['Place_Site'] = df_inscriptions['Place_Site'].str.replace(r'\s+', ' ', regex=True)
df_inscriptions['Place_Site'].unique()
# J'enlève tout ce qui commence par "near" pour n'avoir que des données précises
df_inscriptions = df_inscriptions[~df_inscriptions['Place_Site'].str.startswith('near').fillna(False)]
df_inscriptions['Place_Site'].unique()
```

```python
mapping_corrections = {
    'Aknha': 'Akhna',
    'Dhekhelia': 'Dhekelia',
    'LimassoL': 'Limassol',
    'Rhizokarpasso': 'Rhizokarpaso',
    'Tremethousha': 'Tremetousha',
    'Apheshia': 'Apeshia',
    'Pano Polimidhia': 'Pano Polemidia'
}
df_inscriptions['Place_Site'] = df_inscriptions['Place_Site'].replace(mapping_corrections)
df_inscriptions['Place_Site'].unique()
```

```python
# Supprimer "T." suivi de chiffres, avec ou sans espaces, et éventuellement une lettre (ex: T.A128)
df_inscriptions['Place_Site'] = (df_inscriptions['Place_Site']
                                 .str.replace(r',?\s*T\s*\.?\s*[A-Z]?\d+\w*', '', regex=True)
                                 .str.strip())
print(df_inscriptions['Place_Site'].unique())
df_inscriptions.value_counts('Place_Site')
```

Nettoyage de la colonne Date

```python
df_inscriptions.value_counts('Date')
```

On voit qu'il y a beaucoup de valeurs uniques alors que ça pourrait être résumé très facilement, en mettant tout ce qui est après un retour à la ligne dans une nouvelle colonne par exemple.

```python
split_data = df_inscriptions['Date'].str.split(r'\r', n=1, expand=True)
if 1 in split_data.columns:
    df_inscriptions['Precision_Date'] = split_data[1].str.replace(r'\r', ' ', regex=True).str.strip()
df_inscriptions['Date'] = split_data[0].str.strip()
df_inscriptions['Precision_Date'] = df_inscriptions['Precision_Date'].fillna('')
df_inscriptions.value_counts('Date')
```

```python
print(df_inscriptions['Date'].unique())
```

```python
import numpy as np

periodes_cles = ['Classical', 'Hellenistic', 'Roman', 'Geometric', 'LC', 'Archaic']

df_inscriptions['Date'] = df_inscriptions['Date'].astype(str).replace('nan', '')

def extraire_periode(row_date):
    for p in periodes_cles:
        if p in row_date:
            return pd.Series([p, row_date])
    return pd.Series(["", row_date])
df_inscriptions[['Date', 'Date_Precision_2']] = df_inscriptions['Date'].apply(extraire_periode)
df_inscriptions.loc[df_inscriptions['Date'] == df_inscriptions['Date_Precision_2'], 'Date_Precision_2'] = ""

print(df_inscriptions['Date'].unique())
print(df_inscriptions.value_counts('Date'))
```
