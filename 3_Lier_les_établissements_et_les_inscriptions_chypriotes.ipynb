```python
import geopandas as gpd

gdf_settlement_buffers = gdf_settlement_ecriture_webmercator.buffer(2500) # 2.5km radius

gdf_buffers_to_join = gpd.GeoDataFrame(
    gdf_settlement_ecriture[['ID', 'Toponym', 'Period_large']],
    geometry=gdf_settlement_buffers,
    crs=3857
)

if gdf_inscriptions.crs != gdf_buffers_to_join.crs:
    gdf_inscriptions = gdf_inscriptions.to_crs(gdf_buffers_to_join.crs)

sj = gpd.sjoin(gdf_inscriptions.reset_index(), gdf_buffers_to_join, predicate='within', how='inner')

sj['Date_clean'] = sj['Date'].astype(str).str.strip()
sj['Period_clean'] = sj['Period_large'].astype(str).str.strip()
matched = sj[sj['Date_clean'] == sj['Period_clean']].copy()

counts_by_period = matched.groupby(['ID', 'Period_large']).size().reset_index(name='Num_Inscriptions_Correct')

if 'Num_Inscriptions' in gdf_settlement_ecriture.columns:
    gdf_settlement_ecriture = gdf_settlement_ecriture.drop(columns=['Num_Inscriptions'])

gdf_settlement_ecriture = gdf_settlement_ecriture.merge(
    counts_by_period,
    on=['ID', 'Period_large'],
    how='left'
)
gdf_settlement_ecriture['Num_Inscriptions'] = gdf_settlement_ecriture['Num_Inscriptions_Correct'].fillna(0).astype(int)
gdf_settlement_ecriture = gdf_settlement_ecriture.drop(columns=['Num_Inscriptions_Correct'], errors='ignore')


gdf_settlement_ecriture = gpd.GeoDataFrame(gdf_settlement_ecriture, geometry='geometry', crs='EPSG:4326')

gdf_clean = gdf_settlement_ecriture.dissolve(
    by=['Toponym', 'Period_large'],
    aggfunc={'Num_Inscriptions': 'first', 'ID': 'first'}
).reset_index()

gdf_clean[['Toponym', 'Period_large', 'Num_Inscriptions']].sort_values(by='Num_Inscriptions', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import contextily as cx

CHYPRE_X = (3580000, 3860000)
CHYPRE_Y = (4100000, 4260000)

def calc_size(n):
    return 40 + (np.sqrt(n) * 20)

couleurs = {
    "Bronze Final": "#8e44ad", "GÃ©omÃ©trique": "#2980b9", "ArchaÃ¯que": "#27ae60",
    "Classique": "#f39c12", "HellÃ©nistique": "#d35400", "Romain": "#c0392b"
}
```

```python
epoque_cible = "Bronze Final"
seuil_nom = 70

fig, ax = plt.subplots(figsize=(14, 12), facecolor='white')
color = couleurs.get(epoque_cible, "blue")

ax.set_xlim(CHYPRE_X)
ax.set_ylim(CHYPRE_Y)
cx.add_basemap(ax, source=cx.providers.CartoDB.PositronNoLabels, zorder=0)

data_plot = gdf_clean[gdf_clean['Period_display_name'] == epoque_cible].copy()

if not data_plot.empty:
    data_plot_web = data_plot.to_crs(epsg=3857)

    ax.scatter(
        data_plot_web.geometry.centroid.x,
        data_plot_web.geometry.centroid.y,
        s=calc_size(data_plot_web['Num_Inscriptions']),
        alpha=0.7, color=color, edgecolors='black', linewidth=0.6, zorder=2
    )

    for idx, row in data_plot_web.iterrows():
        if row['Num_Inscriptions'] >= seuil_nom:
            centroid = row.geometry.centroid
            ax.text(
                centroid.x, centroid.y + 3000,
                row['Toponym'],
                fontsize=9,
                fontweight='bold',
                ha='center',
                zorder=4,
                bbox=dict(facecolor='white', alpha=0.5, edgecolor='none', pad=0.1)
            )

    v_min = int(data_plot_web['Num_Inscriptions'].min())
    v_max = int(data_plot_web['Num_Inscriptions'].max())
    v_med = int((v_min + v_max) / 2)

    paliers = sorted(list(set([v_min, v_med, v_max])))

    handles = [ax.scatter([], [], c='gray', alpha=0.3, s=calc_size(v), edgecolors='black') for v in paliers]
    ax.legend(handles, [f"{v} inscr." for v in paliers],
              title=f"Ã‰chelle (Max Ã©q: {v_max})", loc='upper left', frameon=True)

ax.set_title(f"Distribution Ã‰pigraphique : {epoque_cible}", fontsize=20, fontweight='bold', color=color, pad=20)
ax.set_axis_off()

plt.show()
```
